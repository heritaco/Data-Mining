```python
#importar librerias 

import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1 leer 
datos = pd.read_csv('nayarit_elecciones.csv', index_col=0,header =[0,1])
#ver 
datos
```

```python
#ver las columnas 
datos.columns
```

```python
#guardar las columnas en una variable
col_dip = datos ['Diputados'].columns
#encontar por coincidencia de caracteres las columnas de candidatos independientes 
cidip = [col for col in col_dip if col[:8]=="CAND_IND"]
#sumar en eje 1 estas columnas 
dipind = datos['Diputados'] [cidip].sum(axis=1)
#agregar nueva colmna al data frame 
datos [('Diputados','TOTAL IND')] = dipind


#mismo proceso para presidente 
col_pres = datos ['Presidente'].columns
cipres = [col for col in col_dip if col[:8]=="CAND_IND"]
presind = datos['Presidente'] [cipres].sum(axis=1)
datos [('Presidente','TOTAL IND')] =  presind

datos.columns
```

```python
plt.style.use('_mpl-gallery')

# Lo vamos a usar para despues
partido = "PAN"
diputados = datos[("Diputados", partido)]
presidentes = datos[("Presidente", partido)]

# size and color:
sizes = np.random.uniform(15, 80, len(diputados))
colors = np.random.uniform(15, 80, len(diputados))
# - Matplotlib
```

```python
def grafica_votos(partido):
    # Conseguimos los porcentajes de votos para diputados y presidente
    porcentaje_dip = datos[("Diputados", partido)] / datos[("Diputados", "TOTAL_VOTOS")] * 100
    porcentaje_pres = datos[("Presidente", partido)] / datos[("Presidente", "TOTAL_VOTOS")] * 100

    # Conseguimos el máximo de los porcentajes, para la linea
    max_dip = porcentaje_dip.max()
    max_pres = porcentaje_pres.max()
    maxmax = max(max_dip, max_pres)

    # Graficamos
    fig, ax = plt.subplots(figsize=(5, 5)) 
    ax.set_aspect('equal')  # Para que los ejes tengan la misma escala

    ax.scatter(porcentaje_dip, porcentaje_pres, c=colors, vmin=0, vmax=100, alpha=0.8) # - Matplotlib

    plt.plot([0,maxmax], [0, maxmax], color='black', linestyle='--', linewidth=1) # Una linea que empieza en 0 y termina en el máximo
    ax.set_xlabel("% Votos para diputados") # Etiqueta del eje x
    ax.set_ylabel("% Votos para presidente")
    ax.set_title(f"Votos para {partido} en Nayarit \n en {len(diputados)} secciones ") # Título de la gráfica
    
    plt.show() # Mostrar la gráfica
```

```python
grafica_votos("PAN")
grafica_votos("PRI")
grafica_votos("MORENA")
```
