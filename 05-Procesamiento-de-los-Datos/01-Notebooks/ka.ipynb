```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


# Sample data
data = np.array([
    ['Enfermedad', 'p-valor'],
    ['Diabetes', 0.0000001],
    ['Hipertension', 0.0000001],
    ['Infarto', 0.0000001],
    ['Angina de Pecho', 0.0000001],
    ['Insuficiencia Cardiaca', 0.0000001],
    ['Embolia', 0.0000001],
    ['Infeccion de Vas Urinarias', 0.0000001],
    ['Calculos Renales', 0.0000001],
    ['Insuficiencia Renal', 0.0000001]

])

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 8))

# Hide the axes
ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)
ax.set_frame_on(False)

# Create the table
table = ax.table(cellText=data, cellLoc='center', loc='center')

# Customize the table
table.auto_set_font_size(False)
table.set_fontsize(14)
table.scale(1.2, 1.2)

# Apply the same color scheme as the heatmap
cmap = sns.color_palette("Blues_r", as_cmap=True)
for i in range(1, len(data)):
    cell_value = float(data[i][1])
    color = cmap(cell_value)
    for j in range(len(data[i])):
        table[(i, j)].set_facecolor(color)
        table[(i, j)].set_edgecolor('white')  # Set the edge color to white

# Save the figure
plt.savefig('tablap.svg')

# Show the plot
plt.show()
```
