### **Agrupa los datos por sección usando .groupby(). Asegura que los votos por un mismo partido o coalición se sumen. En caso de  manejar variables de texto puedes usar la función de agregación "first".**

#### *Solo para números*

Vamos a ver si son los mismos partidos

````python
p_estado.columns
````

````python
d_estado.columns
````

Son los mimsos partidos

````python
partidos = [
        'PAN', 'PRI', 'PRD', 'PVEM', 'PT', 'MC', 'NA', 'MORENA',
        'ES', 'PAN_PRD_MC', 'PAN_PRD', 'PAN_MC', 'PRD_MC', 'PRI_PVEM_NA',
        'PRI_PVEM', 'PRI_NA', 'PVEM_NA', 'PT_MORENA_ES', 'PT_MORENA', 'PT_ES',
        'MORENA_ES',  'CAND_IND1', 'CAND_IND2']
````

````python
p_estado[partidos].dtypes
````

````python
d_estado[partidos].dtypes
````

Todos son floats

````python
p_estado_sum = p_estado.groupby("CLAVE").sum()
````

````python
p_estado_sum = p_estado_sum[partidos]
p_estado_sum.head(2)
````

````python
votos = p_estado_sum[partidos].sum(axis=0)
total = votos.sum()
porcentaje = votos / total * 100
porcentaje.round(2).sort_values(ascending=False)
````

````python
d_estado_sum = d_estado.groupby("CLAVE").sum()
````

````python
d_estado_sum = d_estado_sum[partidos]
d_estado_sum.head(2)
````

````python
votos = d_estado_sum[partidos].sum(axis=0)
total = votos.sum()
porcentaje = votos / total * 100
porcentaje.round(2).sort_values(ascending=False)
````

Si lo hacemos asi perdemos un poquito de info porque no conocemos los nombres.

#### *Numeros y cadenas*

````python
d_estado.columns
````

Queremos guardan el nombre de estado y municipio para identificar la clave. Luego vamos a usar el Total de votos

````python
info_que_quieremos = ["NOMBRE_ESTADO", "MUNICIPIO"] + partidos + ["TOTAL_VOTOS"]
````

````python
d_estado_sum = d_estado.groupby("CLAVE").sum() # Agrupamos por clave y sumamos los votos
d_estado_sum = d_estado_sum[info_que_quieremos]# Seleccionamos las columnas que queremos 
d_estado_sum.head(2)
````

Para evitar que diga OAXACAOAXACAOAXACAOAXACAOAXACAOAXACAOAXACAOAXA tenemos que hacer un diccionario

#### *Diccionario*

````python
funciones_agg_dip = {}  # Creamos un diccionario 

for col in p_estado.columns:
    if p_estado[col].dtype == "object":
        funciones_agg_dip[col] = "first"
    else:
        funciones_agg_dip[col] = "sum"
````

````python
funciones_agg_dip 
````

Si la columna es de tipo object, se queda con el primer valor, si no, suma los valores

#### *Diccionario de compresión*

Es más rápido

````python
mejor_diccionario_d = { 
    col: "first" if d_estado[col].dtype == "object" else "sum" for col in d_estado.columns
}

mejor_diccionario_p = {
    col: "first" if p_estado[col].dtype == "object" else "sum" for col in p_estado.columns
}
````

````python
p_grouped = p_estado.groupby("CLAVE").agg(mejor_diccionario_p)
p_grouped.head(2)
````

Se repite ```"CLAVE"```

````python
p_grouped = p_grouped.drop("CLAVE", axis=1)
p_grouped.head(2)
````

Solo agarramos la info que queremos

````python
p_grouped = p_grouped[info_que_quieremos]
p_grouped.head(2)
````

Lo mismo para diputados

````python
d_grouped = d_estado.groupby("CLAVE").agg(mejor_diccionario_d)
d_grouped = d_grouped[info_que_quieremos] # Aqui no habia puesto clave
````

### **Une ambas tablas usando los métodos .join() o .merge() y la clave creada en el punto 3.**

Merge es una funcion de pandas que nos permite unir dos dataframes  
how = "inner" es el tipo de merge que queremos hacer, en este caso es un inner join  
on = "CLAVE" es la columna por la que queremos hacer el merge

````python
merge = pd.merge(p_grouped, d_grouped, how="inner", on="CLAVE") 
````

````python
merge.head(2)
````

### **Diferencia los resultados electorales por cada tipo de elección (Diputación, Presidencia) usando MultiIndex en las columnas.**

````python
merge.columns
````

Como hicimos el merge nos quedaron los nombres de las columnas como _x y _y, se los tenemos que quitar

````python
pd.MultiIndex.from_tuples([
    ("Diputados", col[:-2]) # Una tupla con dos elementos, el primero es el nombre de la tabla y el segundo es el nombre de la columna
        if                  # col[:-2] significa que se va a quitar los dos ultimos caracteres
            col [-1:]=="x"  # Si el ultimo caracter es x
        else 
            ("Presidencia", col[:-2])
    for col in merge.columns# Por cada columna en merge
])
````

Le asignamos a ```merge.columns``` el MultiIndex

````python
merge.columns = pd.MultiIndex.from_tuples([  # Se le asigna a merge.columns un MultiIndex
    ("Diputados", col[:-2]) # Una tupla con dos elementos, el primero es el nombre de la tabla y el segundo es el nombre de la columna
        if                  # col[:-2] significa que se va a quitar los dos ultimos caracteres
            col [-1:]=="x"  # Si el ultimo caracter es x
        else 
            ("Presidencia", col[:-2])
    for col in merge.columns# Por cada columna en merge
])
````

Así se ve

````python
merge[[("Presidencia", "PAN"), ("Diputados", "PAN")]] 
````

````python
merge.head(2)
````

Nos quedamos con las secciones que tengan más de 0 votos

````python
merge = merge[merge["Diputados"]["TOTAL_VOTOS"]>0]
merge = merge[merge["Presidencia"]["TOTAL_VOTOS"]>0]
````
