```python
import os
import pandas as pd

# Define the directory containing the xlsx files
directory = r'C:\Heri\GitHub\.Semester\Data-Mining\03-Tasas-de-Interes-al-Consumo\Datos\Series'
directory2 = r'C:\Heri\GitHub\.Semester\Data-Mining\03-Tasas-de-Interes-al-Consumo\Datos\Series-csv'

# Iterate over all files in the directory
for filename in os.listdir(directory):
    file_path = os.path.join(directory, filename) 
    
    # Read the xlsx file
    df = pd.read_excel(file_path)
    
    # Construct the csv file path
    csv_file_path = os.path.join(directory2, filename.replace('.xlsx', '.csv'))
    
    # Save the dataframe to csv
    df.to_csv(csv_file_path, index=False)
```
