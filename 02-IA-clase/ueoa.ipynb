Escribe una función que construya una gráfica de dispersión del porcentaje de votos (respecto al total de votos por sección por tipo de elección) de un partido o coalición en ambas elecciones:  
   - La entrada de la función debe ser el nombre del Partido o la coalición  
   - Los ejes de la gráfica deben especificar a que elección corresponden ("Diputación" o "Presidencia")  
   - Ambos ejes deben estar en el rango 0, 100  
   - El título de la gráfica debe contener la siguiente información: Nombre del partido, nombre de la entidad y número de secciones (puntos) en la gráfica.  
   - Agrega una línea punteada que cruce el gráfico en diagonal del punto (0,0) al punto (0,100)  


**Recomendaciones:**

- Usa fig, ax = plt.subplots() para crear el entorno de la gráfica.  
- Usa plt.plot() para construir la línea recta, y adecuar los parámetros para que sea una línea punteada.  
- Usa ax.set_xlabel para especificar el nombres de los ejes.  
- Personaliza el color de los puntos.  
- Usa "  ax.set_aspect("equal")" para que la relación entre tus ejes sea simétrica.  
- Usa el decorado f'Título' al especificar el título de la gráfica, para poder usar el salto de línea definido por "\n".  


```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

partido = "PAN"


diputados = datos[("Diputados", partido)]
presidentes = datos[("Presidente", partido)]

# size and color:
sizes = np.random.uniform(15, 80, len(diputados))
colors = np.random.uniform(15, 80, len(diputados))


def grafica_dispersion(partido):

    porcentaje_dip = datos[("Diputados", partido)] / datos[("Diputados", "TOTAL_VOTOS")] * 100
    porcentaje_pres = datos[("Presidente", partido)] / datos[("Presidente", "TOTAL_VOTOS")] * 100

    max_dip = porcentaje_dip.max()
    max_pres = porcentaje_pres.max()
    maxmax = max(max_dip, max_pres)

    fig, ax = plt.subplots(figsize=(5, 5))
    ax.set_aspect('equal') 

    ax.scatter(porcentaje_dip, porcentaje_pres, c=colors, vmin=0, vmax=100, alpha=0.8) # Matplotlib

    plt.plot([0,maxmax], [0, maxmax], color='black', linestyle='--', linewidth=1)
    ax.set_xlabel("porcentaje de votos para diputados")
    ax.set_ylabel("porcentaje de votos para presidente")
    ax.set_title(f"Votos para {partido} en Nayarit \n en {len(diputados)} secciones ")
    
    plt.show()
```
