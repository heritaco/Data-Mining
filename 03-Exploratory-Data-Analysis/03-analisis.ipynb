```python
maxrelation_dict = {}

for col in df.columns:
    relationpearson = corr_matrix_pearson[col].abs().sum()
    relationspearman = corr_matrix_spearman[col].abs().sum()
    relationkendall = corr_matrix_kendall[col].abs().sum()

    # Determine the method with the maximum sum
    if relationpearson >= relationspearman and relationpearson >= relationkendall:
        max_method = "pearson"
    elif relationspearman >= relationpearson and relationspearman >= relationkendall:
        max_method = "spearman"
    else:
        max_method = "kendall"

    maxrelation_dict[col] = max_method

maxrelation_dict
```

```python
from statsmodels.tsa.stattools import adfuller

# Evaluar estacionariedad para la serie "G"
result_g = adfuller(data["G"].dropna())
print("Serie G:")
print("Estadístico ADF:", result_g[0])
print("p-valor:", result_g[1])

# Evaluar estacionariedad para la otra serie
serie_idx = "Índice de precios al consumidor, por objeto del gasto, Nacional, Índice general"
result_idx = adfuller(data[serie_idx].dropna())
print("\nSerie", serie_idx, ":")
print("Estadístico ADF:", result_idx[0])
print("p-valor:", result_idx[1])
```

```python
# Crear una copia del DataFrame original
data_stationary = data.copy()

# Diferenciar las series para hacerlas estacionarias
data_stationary.loc[:, "diff_G"] = data_stationary["G"].diff()
data_stationary.loc[:, "diff_IPC"] = data_stationary["Índice de precios al consumidor, por objeto del gasto, Nacional, Índice general"].diff()

# Eliminar los valores NaN generados por la diferenciación
data_stationary = data_stationary.dropna()
```

```python
adfuller(data_stationary.loc[:, "diff_G"])
```

```python
adfuller(data_stationary.loc[:, "diff_IPC"])
```

```python
def make_stationary(series, p_value_threshold=0.05, max_diff=10):
    d = 0
    current_series = series.copy().dropna()
    adf_result = adfuller(current_series)
    p_value = adf_result[1]
    print(f"Diferenciación {d}: p-valor = {p_value}")
    
    # Se diferencia recursivamente hasta cumplir la condición o alcanzar el máximo permitido
    while p_value >= p_value_threshold and d < max_diff:
        d += 1
        current_series = current_series.diff().dropna()
        adf_result = adfuller(current_series)
        p_value = adf_result[1]
        print(f"Diferenciación {d}: p-valor = {p_value}")
        
    return current_series, d, adf_result
```

```python
# Aplicar la función a la serie 'G'
g, num_diff, adf_res = make_stationary(data["G"])
print("Número de diferenciaciones realizadas:", num_diff)
print("p-valor final:", adf_res[1])
```

```python
# Aplicar la función a la serie 'G'
ipc, num_diff, adf_res = make_stationary(data["Índice de precios al consumidor, por objeto del gasto, Nacional, Índice general"])
print("Número de diferenciaciones realizadas:", num_diff)
print("p-valor final:", adf_res[1])
```

```python
from sklearn.linear_model import LinearRegression

# Combinar las series estacionarias en un DataFrame
df_stationary = pd.DataFrame({
    "stationary_y": g,
    "stationary_ipc": ipc
}).dropna()

# Crear dos lags para la serie independiente estacionaria
df_stationary.loc[:, "stationary_ipc_Lag1"] = df_stationary["stationary_ipc"].shift(1)
df_stationary.loc[:, "stationary_ipc_Lag2"] = df_stationary["stationary_ipc"].shift(2)
df_stationary = df_stationary.dropna()

# Definir la variable dependiente y la matriz de variables independientes
X = df_stationary[["stationary_ipc_Lag1", "stationary_ipc_Lag2"]]
y = df_stationary["stationary_y"]

# Ajustar el modelo de regresión lineal
model = LinearRegression()
model.fit(X, y)

```

```python
# Imprimir los coeficientes e intercepto
print("Coeficientes:", model.coef_)
print("Intercept:", model.intercept_)
print("R^2:", model.score(X, y))
```

# VAR

```python
from statsmodels.tsa.api import VAR

df_var_diff = pd.DataFrame({
    "stationary_y": g,
    "stationary_ipc": ipc
}).dropna()

# Ajustar el modelo VAR usando un máximo de 2 lags y seleccionando el orden óptimo según AIC
model_var = VAR(df_var_diff)
results_var = model_var.fit(maxlags=3, ic='aic')

# Imprimir el resumen del modelo VAR
print(results_var.summary())
```

### **Interpretación del VAR**

Este modelo de **Vectores Autorregresivos (VAR)** estima la relación entre dos series de tiempo: **stationary_y** y **stationary_ipc**, utilizando tres rezagos. Aquí desglosamos los resultados clave:

---

### **1. Evaluación Global del Modelo**
- **Número de observaciones:** 150  
- **Número de ecuaciones:** 2 (una para cada variable dependiente)
- **Criterios de Información:**  
  - **AIC:** 11.1880  
  - **BIC:** 11.4690  
  - **HQIC:** 11.3022  
  Estos valores indican la calidad del ajuste; un valor menor es preferible, pero su interpretación es relativa.

- **Log-likelihood:** -1250.78  
  Este valor es útil para comparar modelos alternativos; un valor mayor indicaría mejor ajuste.

- **Determinante de la matriz de varianzas-covarianzas de los errores (\(\Omega\)):** 65968.6  
  Una menor varianza de los errores indicaría mejor ajuste.

---

### **2. Interpretación de la Ecuación para `stationary_y`**
\[
\text{stationary\_y}_t = c + \sum_{i=1}^{3} \beta_i \text{stationary\_y}_{t-i} + \sum_{i=1}^{3} \gamma_i \text{stationary\_ipc}_{t-i} + \epsilon_t
\]
- **Intercepto:** 27.08 (no significativo, \( p = 0.613 \))  
  No hay evidencia de una tendencia constante en la serie.

- **Efectos rezagados de `stationary_y`:**  
  - \( L1: -0.019 \) (no significativo, \( p = 0.799 \))  
  - \( L2: 0.0188 \) (no significativo, \( p = 0.797 \))  
  - \( L3: -0.4438 \) (**muy significativo, \( p = 0.000 \)**)  

  Solo el tercer rezago es relevante, lo que sugiere que `stationary_y` tiene una dependencia con su propio valor hace 3 periodos.

- **Efectos rezagados de `stationary_ipc`:**  
  - \( L1: -289.18 \) (**significativo, \( p = 0.034 \)**)  
  - \( L2: -337.00 \) (**significativo, \( p = 0.025 \)**)  
  - \( L3: -7.38 \) (no significativo, \( p = 0.958 \))  

  `stationary_ipc` afecta negativamente a `stationary_y` con un rezago de 1 y 2 periodos, pero el efecto desaparece en el tercer rezago.

---

### **3. Interpretación de la Ecuación para `stationary_ipc`**
\[
\text{stationary\_ipc}_t = c + \sum_{i=1}^{3} \beta_i \text{stationary\_y}_{t-i} + \sum_{i=1}^{3} \gamma_i \text{stationary\_ipc}_{t-i} + \epsilon_t
\]
- **Intercepto:** 0.0065 (no significativo, \( p = 0.840 \))  
  No hay evidencia de una tendencia constante.

- **Efectos rezagados de `stationary_y`:**  
  - Todos los coeficientes son pequeños y no significativos, excepto \( L2: 0.000081 \) (\( p = 0.065 \)), que está cerca del umbral de significancia.

  Esto sugiere que `stationary_y` apenas influye en `stationary_ipc`.

- **Efectos rezagados de `stationary_ipc`:**  
  - \( L1: -0.5010 \) (**muy significativo, \( p = 0.000 \)**)  
  - \( L2: -0.2456 \) (**significativo, \( p = 0.007 \)**)  
  - \( L3: -0.2344 \) (**significativo, \( p = 0.005 \)**)  

  `stationary_ipc` muestra una fuerte dependencia de sus valores pasados, lo que indica que sigue un proceso autorregresivo importante.

---

### **4. Matriz de Correlación de los Errores**
\[
\begin{bmatrix}
1.0000 & -0.0598 \\
-0.0598 & 1.0000
\end{bmatrix}
\]
- La correlación entre los residuos de ambas ecuaciones es **-0.0598**, cercana a 0, lo que sugiere que los choques en `stationary_y` y `stationary_ipc` son prácticamente independientes.

---

### **Conclusiones**
1. `stationary_y` es fuertemente influenciado por su tercer rezago y por los valores rezagados de `stationary_ipc` (con 1 y 2 periodos de retraso).
2. `stationary_ipc` tiene un comportamiento autorregresivo fuerte y apenas responde a cambios en `stationary_y`.
3. La baja correlación en los errores sugiere que los choques de ambas variables no están altamente relacionados.

Este modelo sugiere que `stationary_ipc` es un predictor más relevante para `stationary_y`, pero no viceversa. Para mejorar el ajuste, podrías probar diferentes rezagos o métodos de selección de modelos como AIC/BIC.

Lo interpreté observando la **significancia estadística** de los coeficientes en ambas ecuaciones:  

1. En la ecuación de **`stationary_y`**, los coeficientes de los rezagos de `stationary_ipc` (L1 y L2) son **estadísticamente significativos** con \( p = 0.034 \) y \( p = 0.025 \), respectivamente. Esto indica que `stationary_ipc` tiene un impacto en `stationary_y`.  

2. En la ecuación de **`stationary_ipc`**, los coeficientes de los rezagos de `stationary_y` son **prácticamente nulos** y no significativos (los valores \( p \) están muy por encima de 0.05, excepto L2 que está cerca con \( p = 0.065 \)). Esto sugiere que `stationary_y` **no afecta significativamente** a `stationary_ipc`.  

Dado que `stationary_ipc` parece influir en `stationary_y`, pero no al revés, concluí que `stationary_ipc` es un predictor más relevante para `stationary_y`, mientras que `stationary_y` no es un predictor fuerte para `stationary_ipc`.
